## Model

Three models were applied to the data:
1. Linear Regression Base Model
2. Random Forest Model
3. Tuned Random Forest Model

I will use the Tuned Random Trees model for the purpose of solving this problem. This is because it is able to predict the sales values with more accuracy than the others.

------------------------------------------------------------

- RMSE = 1,060.790
- R^2 = 0.592

This model returns the highest accuracy prediction ratio of 0.592 (59.2%), signifying that this is the model that will be used.

Interpreting the metrics:

-- MAE (737.28) - this is the Mean Average Error, signifying that on average the model will predict a sales value with a difference of $737.28 from the actual value

-- RMSE (1060.79) - this is the root of the Mean Squared Error, and if it is significantly different from the MAE it signifies that there are some large errors which are being disproportionately applied. This implies that the data could have outliers which need to be addressed.

-- R^2 (0.592) - this is the percentage of the predicted value that can be directly attributable to the data. A percentage of 59.2% is good, but it could be better.

## Recommendations:

This model is overfit, a train score of .715 versus a test score of .592. This suggests that there is room for additional parameter tuning to improve the test score
